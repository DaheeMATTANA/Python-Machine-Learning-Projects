##### Redundant Column (ID column)

```python
df = df.drop('id', axis = 1)
```

##### Datetime object column

```python
df['date']     # Currently string
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['date']
```

##### Extract year and month

```python
df['year'] = df['date'].apply(lambda date : date.year)
df['month'] = df['date'].apply(lambda date : date.month)
```

```python
df = df.drop('date', axis = 1)
```

##### Zip Code Column

```python
df['zipcode'].value_counts()

# Encoding 70 categories will be a too much
# In real cases, use the domaine knowledge to deal with this kind of data (such as grouping by south, north zipcode)
```

```python
# in this case study, we will remove it
df = df.drop('zipcode', axis = 1)
```

##### Renovation year column : encode into categorical value

```python
df['yr_renovated'].value_counts()    
# year 0 means : not renovated, It makes more sense to encode this into 0 : not renovated, 1 : renovated
```

```python
def yr_renov_encode(yr) :
    if yr != 0 :
        return 1
    else :
        return 0
```

```python
df['yr_renovated'] = df['yr_renovated'].apply(lambda yr : yr_renov_encode(yr))
```

##### Renovation year column :  Dummy variables

```python
renovated = pd.get_dummies(df['yr_renovated'], drop_first = True, dtype = int)
renovated.columns = ['renovated']
```

```python
df = pd.concat([df, renovated], axis = 1)
df.drop('yr_renovated', axis = 1, inplace = True)
```

##### Sqft Basement column : Encode into categorical value

```python
df['sqft_basement'].value_counts()
# sqft 0 means : no basement, It makes more sense to encode this into 0 : no basement, 1 : there is a basement
```

```python
df['sqft_basement'] = df['sqft_basement'].apply(lambda bm : yr_renov_encode(bm))
```

#####  Sqft Basement column : Dummy variables

```python
basement = pd.get_dummies(df['sqft_basement'], drop_first = True, dtype = int)
basement.columns = ['basement']
```

```python
df = pd.concat([df, basement], axis = 1)
df.drop('sqft_basement', axis = 1, inplace = True)
```

## 1/ ANN Regression

##### Set Split

```python
X = df.drop('price', axis = 1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

##### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

##### Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))

model.add(Dense(1))

model.compile(optimizer = 'adam', loss = 'mse')
```

```python
model.fit(x = X_train, y = y_train, validation_data = (X_test, y_test), batch_size = 128, epochs = 400, verbose = 0)
```

##### Losses

```python
losses = pd.DataFrame(model.history.history)
```

```python
# Compare loss on training, on test validation to see if we are not overfitting
losses.plot()
plt.show()
```

```python
# This is good, there is no sudden increase in validation data 
```

##### Prediction

```python
predictions = model.predict(X_test)
```

##### Evaluation

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
np.sqrt(mean_squared_error(y_test, predictions))     # Root mean squared error
```

```python
mean_absolute_error(y_test, predictions)
```

```python
df['price'].describe()
```

```python
5.402966e+05       # We are off by 20% (mean of actual price vs mean_absolute error)
```

```python
explained_variance_score(y_test, predictions)
```

```python
plt.figure(figsize = (10, 6))
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, 'r')    # Perfect prediction line

plt.title('Actual Value vs Predictions')
plt.show()
```

##### Error Distribution

```python
errors = y_test.reshape(6480, 1) - predictions

sns.distplot(errors)
plt.show()
```
